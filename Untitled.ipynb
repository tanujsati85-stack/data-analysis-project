```python
import cv2
import face_recognition
import numpy as np

# Load reference image
reference_img = face_recognition.load_image_file(r"C:\Users\Admin\Downloads\WhatsApp Image 2025-09-11 at 22.05.37.jpeg")
reference_encoding = face_recognition.face_encodings(reference_img)[0]

# Convert reference image to OpenCV format (for display)
ref_display = cv2.cvtColor(reference_img, cv2.COLOR_RGB2BGR)

# Open the video
video_path = r"C:\Users\Admin\Downloads\WhatsApp Video 2025-09-11 at 22.06.10.mp4"
video_capture = cv2.VideoCapture(video_path)

found_count = 0

while True:
    ret, frame = video_capture.read()
    if not ret:
        break

    # Detect faces in the frame
    face_locations = face_recognition.face_locations(frame)
    face_encodings = face_recognition.face_encodings(frame, face_locations)

    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        match = face_recognition.compare_faces([reference_encoding], face_encoding, tolerance=0.6)
        if match[0]:
            found_count += 1
            # Draw rectangle around matched face
            cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
            cv2.putText(frame, "MATCH", (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 
                        0.9, (0, 255, 0), 2)

    # Resize images for side-by-side display
    ref_resized = cv2.resize(ref_display, (200, 200))
    frame_resized = cv2.resize(frame, (640, 360))

    # Put them together horizontally
    combined = np.hstack((ref_resized, frame_resized))

    cv2.imshow("Photo (Left) | Video (Right)", combined)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()

print(f"✅ Face appeared {found_count} times in the video.")
```
